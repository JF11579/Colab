```python
import pandas as pd
import io
import requests

# URL of the text file
url = 'https://www2.census.gov/programs-surveys/saipe/datasets/2021/2021-school-districts/sd21-ny.txt'

# Use requests to download the file
response = requests.get(url)

# Create a BytesIO object and pass the content of the response object to it
data = io.BytesIO(response.content)

# Use Pandas to read the text data into a DataFrame with a single column
df = pd.read_csv(data, header=None)

# Split the single column into multiple columns based on white space
df = df[0].str.split(expand=True)

# Rename the columns of the new DataFrame
df.columns = ['col1', 'col2', 'col3','four' ,'five', 'six', 'seven', 'eight', 'nine', 'ten',
              'eleven', 'twelve', 'thriteen', 'fourteen', 'fifteen' , 'sixteen' ]

# Display the DataFrame
print(df)
```

```python
df.head(50)
```

Some of the school district names are not 3 words long but four.  concatenating put text into numeric cols.

Instead lets insert a blank column and then relocate it.before the numbers and concatenate that.  Any fourth wor should land there.

```python
# insert empty col btwn cols 6&7

df.insert(6, 'inserted_col' , np.nan)
```

```python
df.head()
```

```python
# making sure that I dont concatenate the wrong col 
# becasue Py starts counting at 0, sometimes.
#. df['six'] = District
```

```python
# Concatenate columns 3 through 6 with a space between words
df['District_Name'] = df.iloc[:, 3:7].apply(lambda x: ' '.join(x.astype(str)), axis=1)

# Display the DataFrame
print(df)
```

```python
df.head(50)
```

```python
# We also need to concatenate cols 9:11

df['Tag_Name_Date'] = df.iloc[:, 11:12].apply(lambda x: ' '.join(x.astype(str)), axis=1)

# Display the DataFrame
print(df)
```

```python
df.head(3)
```

```python
# # Drop some of these columns
# # df = df.drop(columns=['col2', 'col4'])

df = df.drop(columns=['col3', 'four', 'five' , 'six' ,
                      'inserted_col',
                           'ten' , 'eleven', 'twelve','thriteen' , 
                      'fourteen' , 'fifteen' , 'sixteen' ])
```
