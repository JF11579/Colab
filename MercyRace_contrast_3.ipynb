```r
# First get summary stats

Median_scores <- All_long %>%
  group_by( Education_Level ,Race_Ethnicity  ) %>%
    summarise_all(median)

Median_scores   
```

### High School Correlations

```r
High_school_Complt_median  <- Median_scores %>%
                    filter(Education_Level == 'High school completion1')
```

```r
High_school_Complt_median 
```

```r
Asian_HS_cmplt_median <- High_school_Complt_median %>%
                        filter(Race_Ethnicity == 'Asian')
                      
Asian_HS_cmplt_median
```

```r
Black_HS_cmplt_median <- High_school_Complt_median %>%
                        filter(Race_Ethnicity == 'Black')
```

```r
Hispanic_HS_cmplt_median <- High_school_Complt_median %>%
                        filter(Race_Ethnicity == 'Hispanic')
```

```r
White_HS_cmplt_median <- High_school_Complt_median %>%
                        filter(Race_Ethnicity == 'White')
```

```r
# VERY high P value, no statistical significance, just randomness, 
#
# A p-value below 0.05 indicates there is a connection, 
# that it the correlation is unlilkely random chance

kruskal.test(list(Asian_HS_cmplt_median, Black_HS_cmplt_median, Hispanic_HS_cmplt_median , White_HS_cmplt_median))
```
