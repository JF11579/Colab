## STANDARD ERROR OF THE MEAN

Std error of mean = pop std deviation, N^0.5

Prob 99% = 2.576

Probability of 95% - z 1.96

Prob 90% = z = 1.645

Prob 85% = 1.036

Prob 80 =  0.842 

# Get CONFIDENCE INTERVAL 

Conf Inter = 90%

Sample Mean = 129.2

Pop Std Dev = o~ = 40.60 

N = 32

Now Calc for Std error of mean:

o~x = pop std dev / N^0.5

o~ = 40.6/ 32^0.5 = 40.6/5.66 = 7.13



  

```r
# Get Conf Interval  for 90%
# We want the range of numbers that cover 90% of order
# two tail: the lowest and the highest.
#
sample_mean <- 129.2
Pop_Std_Dev <- 40.6
N <- 32
#N_root
#
root <- 32^0.5
root_round <- round(root,2)
Std_Error_of_Mean <- Pop_Std_Dev/root_round
Std_Error_of_Mean
rounded_Std_Error_of_Mean <- round(Std_Error_of_Mean,3)
paste("Standard Error of the Mean is",rounded_Std_Error_of_Mean )
```

## We need to find the critical z score.

We want 90% That is a 5% boundary on each side (5+90+5)

The Significance level for 90% for a 2 tail test means wwe want 95% and then that 95% will be the matching same #.  95% = 1.645

So the left side z score is - 1.645 and the right side = pos 1.645.

Now that we have the z scores we can calculate the upper and and lower conficencce levels.

UCL = sample mean + zscore times standard error of the mean

UCL = 129.2 times + (z score times Sample Error of the mean)

UCL = 129.2 +(1.645*7.173)

```r
z_score <- 1.645
rounded_Std_Error_of_Mean
sample_mean
#
ans <- z_score * rounded_Std_Error_of_Mean
ans_rounded <-round(ans,2)

UCL <- sample_mean + ans_rounded
UCL
paste("The UCL is", UCL)
############
######. THE LCL ###########
LCL <- sample_mean - ans_rounded
LCL
paste("The LCL is", LCL)
```
